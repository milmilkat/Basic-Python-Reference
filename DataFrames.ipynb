```python
#"s":100 if x == 'Low income' else 10 for x in stats.IncomeGroup

sizes = []
colors = []

for i in stats.IncomeGroup:
    if i == 'High income':
        sizes.append(600)
        colors.append("blue")
    elif i == "Upper middle income":
        sizes.append(400)
        colors.append("green")
    elif i == "Lower middle income":
        sizes.append(200)
        colors.append("red")
    else:
        sizes.append(100)
        colors.append("orange")
        
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate',     \
                  fit_reg=False, size= 8, aspect= 1, legend= True,  \
                  scatter_kws= {"s":sizes,"color":colors})
```
